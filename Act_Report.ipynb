#### The most popular dog class: 
These are the various stages of dog: doggo, pupper, puppo and floof(er).<br>
1. Doggo are usually older. 
2. pupper is a small doggo, usually younger.
3. puppo is a transitional phase between pupper and doggo. Jus like a dog equivalent of a teenager.
4. floof(er) is any dog. However, this label is commonly given to dogs with seemingly eccess fur.<br>

From the analysis result, the most common dog stage is pupper with a count of 155. Again, I made use of the value_counts() method to get this result. Also, the bar chart shows the same result for pictorial reference.


#### Relationship between dog ratings and retweet count:
The dogs ratings has an effect on how people or dog lovers reacts to the tweets. Ratings are sometimes given to dogs based on what the dog has been through, passing through at the moment or even what the dog has archived. The rating has almost nothing to do with how cute the dog appears to be. Looking at how the both dog ratings and retweet count relates to each other, the results shows that lower dog ratings between 1/10 to 8/10 has low retweet counts while from 10/10 and above has higher retweet counts.
The relationship between  ratings and retweets shows that dogs with lower ratings has low retweets counts compared to the one's with higher ratings. We can see that only one dog with rating of 13/10 was able to reach remarkable result of over 70000 retweet.
